```python
#ID:2358554
#NAME:SABIN CHAULAGAIN
#BMR
def m_bmr(gender,weight,height,age):
  if gender=="m":
    bmr=88.362+(13.397*weight)+(4.799*height)-(5.677*age)
  elif gender=="f":
    bmr=447.593+(9.247*weight)+(3.098*height)-(4.330*age)
  else:
    print("Invalid Gender")
    return None
  return bmr
#prompt user details
gender=input("enter your gender male or female:")
weight=float(input("enter your weight in kg:"))
height=float(input("enter your height in centimeter:"))
age=int(input("enter your age in present years:"))
#calculate BMR now
bmr=m_bmr(gender,weight,height,age)
if bmr is None:
  exit()
#exercise level now for user
print("exercise category codes:")
print("0:Little to no exercise:")
print("1:Light exercise(1-3 days per week):")
print("2:Moderate exercise(3-5 days per week):")
print("3:Heavy exercise(6-7 days per week):")
print("4:Very heavy exercise (twice per day,extra heavy workout):")
exercise_cate=int(input("Enter your exercise category code now:"))
#calculate daily required calorie for you in daily exercise level
if exercise_cate==0:
  daily_calories=bmr*1.2
elif exercise_cate==1:
  daily_calories=bmr*1.375
elif exercise_cate==2:
  daily_calories=bmr*1.55
elif exercise_cate==3:
  daily_calories=bmr*1.725
elif exercise_cate==4:
  daily_calories=bmr*1.9
else:
  print("Invalid Exercise category code for now:")
  exit()
  #output
print("your BMR is:",bmr)
print("Your daily required calories intake is:",daily_calories)
```

```python
#NAME:SABIN CHAULAGAIN
#2358554
#User recursive low and high values as integers
def recur(low,high):
  if low>high:
    return 0
  elif low==high:
    return low
  else:
    return low+recur(low+1,high)
#user to enter value high low
try:
  low=int(input("Enter low value of student number:"))
  high=int(input("Enter high value of student number:"))
#checking value if low or high
  if low>high:
    print("Error:Low value is greater than high value:")
  else:
  #calling recursive function to calculate sum of integers low and high
    print("Sum of integers between {} and {} is {}.".format(low,high,recur(low,high)))
except ValueError:
  print("Error:input values arenot integer.")
```

```python
import matplotlib.pyplot as plt
import numpy as np
#Defining function
def f(x5):
  return 3*x5**2
def f1(x6):
  return 4*x6-3
#generating 100 evenly space x-value between -5 and 15
x_values=np.linspace(-5,15,100)
#calculate corresponding y-value for f and f(1) using x_values array
y_values_1=f(x_values)
y_values_2=f1(x_values)
#plotting two function on same graph with red and blue triangle
plt.plot(x_values,y_values_1,'ro',label='f(x5)=3x^2')
plt.plot(x_values,y_values_2,'b^',label='f(x6)=4x-3')
#addding title,x-label,y-label grib and legend
plt.title('graph of f(x5)=3x^2 and f(x6)=4x-3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
#displaying graph
plt.show()
```
